```python
# Import the time module for time-related functions
import time

# Import the clear_output function from IPython.display to clear the output of the current cell
from IPython.display import clear_output


# Base URL of the webpage to scrape data from, with a placeholder for the page number
base_url = 'https://www.makaan.com/hyderabad-residential-property/buy-property-in-hyderabad-city?page='

# Initialize empty lists to store the scraped data
owner = []
no_bhk = []
locality = []
price = []
area_sqft = []
con_status = []

# Loop through a range of page numbers from 1 to 10 (inclusive)
for i in range(1, 11):
    # Construct the URL for the current page
    url = base_url + str(i)

    # Print a message indicating which page is being scraped
    print(f"Scraping info from page {i}")

    # Clear the output to keep the display clean
    clear_output(wait=True)

    # Pause for 3 seconds to avoid overwhelming the server
    # time.sleep(3)

    # Send an HTTP GET request to the current page URL and parse the response with BeautifulSoup
    # response = requests.get(url)
    # soup = BeautifulSoup(response.text, 'html.parser')

    # Find all 'span' elements with the attribute 'itemprop' set to 'addressLocality' and append their text content to the locality list
    loc = soup.find_all('span', attrs={'itemprop': 'addressLocality'})
    for i in loc:
        locality.append(i.text)

    # Find all 'span' elements with the class 'seller-type' and append their text content to the owner list
    own = soup.find_all('span', attrs={'class': 'seller-type'})
    for o in own:
        owner.append(o.text)

    # Find all 'a' elements with the class 'typelink' and append the text content of their 'span' child elements to the no_bhk list
    bh = soup.find_all('a', attrs={'class': 'typelink'})
    for b in bh:
        no_bhk.append(b.span.text)

    # Find all 'div' elements with the attribute 'data-type' set to 'price-link' and append their text content to the price list
    prc = soup.find_all('div', attrs={'data-type': 'price-link'})
    for p in prc:
        price.append(p.text)

    # Find all 'td' elements with the class 'size' and append their text content to the area_sqft list
    ar = soup.find_all('td', attrs={'class': 'size'})
    for a in ar:
        area_sqft.append(a.text)

    # Find all 'td' elements with the class 'val' and append their text content to the con_status list
    stat = soup.find_all('td', attrs={'class': 'val'})
    for s in stat:
        con_status.append(s.text)
```

```python
# Create a DataFrame using the pandas library with the specified columns and data
data = pd.DataFrame({
    'Owner': owner,
    'No_Bhk': no_bhk,
    'Locality': locality,
    'Price': price,
    'Area': area_sqft,
    'Cons_status': con_status
})

# Display the DataFrame
data
```
